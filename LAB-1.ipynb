```python
import numpy as np
import matplotlib.pyplot as plt

# Sigmoid function
def sigmoid(x):
    return 1 / (1 + np.exp(-x))

# Tanh function
def tanh(x):
    return np.tanh(x)

# ReLU function
def relu(x):
    return np.maximum(0, x)

# Linear function
def linear(x):
    return x
```

```python
# Test input
x = np.array([-1, 0, 1])
```

```python
print("Sigmoid:", sigmoid(x))
```

```python
print("Tanh:", tanh(x))
```

```python
print("ReLU:", relu(x))
```

```python
print("Linear:", linear(x))
```

```python
x = np.linspace(-10, 10, 100)

y_sigmoid = sigmoid(x)
y_tanh = tanh(x)
y_relu = relu(x)
y_linear = linear(x)

plt.plot(x, y_sigmoid, 'r', label='Sigmoid')
plt.plot(x, y_tanh, 'b', label='Tanh')
plt.plot(x, y_relu, 'g', label='ReLU')
plt.plot(x, y_linear, 'y', label='Linear')

plt.xlabel('Input')
plt.ylabel('Output')
plt.legend(loc='best')

plt.show()
```

```python
import numpy as np

def mean_squared_error(y_pred, y_true):
    return np.mean((y_pred - y_true)**2)

def mean_absolute_error(y_pred, y_true):
    return np.mean(np.abs(y_pred - y_true))

def binary_cross_entropy(y_pred, y_true):
    return -np.mean(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred))

def categorical_cross_entropy(y_pred, y_true):
    return -np.mean(np.sum(y_true * np.log(y_pred)))

def log_loss(y_pred, y_true):
    return -np.mean(np.sum(y_true * np.log(y_pred) + (1 - y_true) * np.log(1 - y_pred)))

def softmax(x):
    exps = np.exp(x)
    return exps / np.sum(exps)
```

```python
y_pred = np.array([0.1, 0.3, 0.6])
y_true = np.array([1, 0, 0])
```

```python
print("Mean Squared Error:", mean_squared_error(y_pred, y_true))
```

```python
print("Mean Absolute Error:", mean_absolute_error(y_pred, y_true))
```

```python
print("Binary Cross-Entropy:", binary_cross_entropy(y_pred, y_true))
```

```python
print("Categorical Cross-Entropy:", categorical_cross_entropy(y_pred, y_true))
```

```python
print("Log-loss:", log_loss(y_pred, y_true))
```

```python
import numpy as np

inputs = np.array([[2, 1, -3], [-2, 2, -3], [0, 3, 0], [1, 4, 2]])
outputs = np.array([-1, 0, 1, 1])

learning_rate = 0.2

weights = np.array([1, 1, 0, -1], dtype=float)

num_iterations = 10

for i in range(num_iterations):
    for j in range(inputs.shape[0]):
        y_pred = np.dot(inputs[j], weights[0:]) #+ weights[0]
        weights[1:] += learning_rate * (outputs[j] - y_pred) * inputs[j]
        #weights[0] += learning_rate * (outputs[j] - y_pred)

print("Final weights:", weights)
```
